```python
import selenium
from selenium import webdriver as wb
from selenium.webdriver.support.ui import Select
import pandas as pd
import time
from tqdm import tqdm
```

```python
#Opening Firefox browser
wbD = wb.Firefox(executable_path=r'C:\Program Files\Mozilla Firefox\geckodriver.exe')
```

```python
#Running loop to store the product links in a list
listOflinks =[]
for i in range(1, 2):
    wbD.get('https://www.trendyol.com/kadin-pantolon-x-g1-c70?pi={}'.format(i))
    time.sleep(4)
    productInfoList = wbD.find_elements_by_class_name('p-card-wrppr')
    for el in productInfoList:
        if (el.text != "" and el.text !="Sponsorlu ürün"):
            try :
                pp2=el.find_element_by_tag_name('a')
                listOflinks.append(pp2.get_property('href'))
            except:
                pass
```

```python
wbD.get(listOflinks[0])
time.sleep(3)
lists = wbD.find_elements_by_class_name('detail-attr-item')
keys = []
values = []
for i in range(len(lists)):
    key = lists[i].find_element_by_xpath('//*[@id="product-detail-app"]/div/section/div/ul/li[{}]/span[{}]'.format(i+1, 1)).text         
    value = lists[i].find_element_by_xpath('//*[@id="product-detail-app"]/div/section/div/ul/li[{}]/span[{}]/b'.format(i+1, 2)).text 
    keys.append(key)
    values.append(value)
```

```python
for i in tqdm(listOflinks):
    wbD.get(i)
    time.sleep(3)
    sku = wbD.find_element_by_class_name('pr-new-br').text
    category= wbD.find_element_by_id('marketing-product-detail-breadcrumb').text
    product_description = wbD.find_element_by_class_name('info-wrapper').text
    lists = wbD.find_elements_by_class_name('detail-attr-item')
    keys = []
    values = []
    for j in range(len(lists)):
        key = lists[j].find_element_by_xpath('//*[@id="product-detail-app"]/div/section/div/ul/li[{}]/span[{}]'.format(j+1, 1)).text         
        value = lists[j].find_element_by_xpath('//*[@id="product-detail-app"]/div/section/div/ul/li[{}]/span[{}]/b'.format(j+1, 2)).text 
        keys.append(key)
        values.append(value)
    temp ={
        'SKU':sku,
        'Category':category,
        'Product_Description':product_description,
        'keys': keys,
        'values': values,
        'linkofproduct':i}
    alldetails.append(temp)
```

```python
df = pd.DataFrame(alldetails)
```
